Bar Plot

```python

import numpy as np 
import matplotlib.pyplot as plt 
  
X = ['Taskset 1 (Total Util: x)','Taskset 2 (Total Util: x)','Taskset 3 (Total Util: x)','Taskset 4 (Total Util: x)', 'Taskset 5 (Total Util: x)']
WFD = [10,20,20,40,20]
BFD = [20,30,25,30,30]
LST = [10,20,30,40,50]

X_axis = np.arange(len(X))
plt.figure(figsize=(20, 10))  
plt.bar(X_axis - 0.2, WFD, 0.2, label = 'WFD', color=(0.9, 0.2, 0.1, 0.4))
plt.bar(X_axis + 0.0, BFD, 0.2, label = 'BFD', color=(0.2, 0.9, 0.6, 0.4))
plt.bar(X_axis + 0.2, LST, 0.2, label = 'LST', color=(0.0, 0.3, 0.9, 0.4))
  
plt.xticks(X_axis, X)
plt.xlabel("Groups", fontsize=20)
plt.ylabel("Number of Students", fontsize=20)
plt.tick_params(labelsize=10)
plt.title("Average Energy - Total Task Utilization", fontsize=20)
# plotting a line plot after changing it's width and height



plt.legend()
plt.show()
```
